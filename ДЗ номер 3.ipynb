```python

# -*- coding: utf-8 -*-
import matplotlib.pyplot as plt
import numpy as np

# от 0.1 до 1000 сделать 10000 точек.Чем больше точек, тем точнее график.
x = np.linspace(0.1, 1000, 10000)

# y1...y6 - линии одного графика, для каждой линии свой набор точек.
y1 = 2.72**(-0.1*x)
y2 = 2.72**(-0.1*x) + 0.01
y3 = 2.72**(-0.01*x) + 0.1
y4 = 2.72**(-0.002*x) 
y5 = np.cos(x/792) + 0.3
y6 = np.cos((x-10)/600)

# будет 1 график, на нем 6 линии:
fig, ax = plt.subplots()
# функция y1(x), синий, надпись y(x)
ax.plot(x, y1, color="blue", label="голубая линия")
ax.plot(x, y2, color="red", label="оранжевая линия")
ax.plot(x, y3, color="green", label="зеленая линия")
ax.plot(x, y4, color="black", label="black")
ax.plot(x, y5, color="orange", label="orange")
ax.plot(x, y6, color="pink", label="pink")
# подпись у горизонтальной оси х
ax.set_xlabel("x")
# подпись у вертикальной оси y
ax.set_ylabel("y")
# показывать условные обозначения
ax.legend()

'''
записываем график в картинку png
'''
#показать рисунок
plt.show()
# сохранить в файл 1.png
#fig.savefig('1.png')
```
